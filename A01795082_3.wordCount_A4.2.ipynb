```python
import os, time  # Importaciones en la misma línea (mala práctica)

def ReadFile(filePath):
    """Lee palabras de un archivo y maneja errores en los datos."""
    words = []
    try:
        with open(filePath, "r", encoding="utf-8") as f:
            for line in f:
                words.extend(line.strip().split())
    except FileNotFoundError as e:
        print(f"Error: No se encontró el archivo {filePath} -> {e}")
        return None
    except Exception as e:
        print(f"Error al leer el archivo {filePath} -> {e}")
        return None
    return words

def CountWords(wordList):
    """Cuenta las palabras en la lista y devuelve un diccionario."""
    counts = {}
    for word in wordList:
        counts[word] = counts.get(word, 0) + 1
    return counts

def GetTop20(wordCounts):
    """Obtiene las 20 palabras más frecuentes."""
    return sorted(wordCounts.items(), key=lambda x: x[1], reverse=True)[:20]

def WriteResults(outputPath, wordCounts):
    """Escribe los resultados en un archivo y los imprime en pantalla."""
    top_20 = GetTop20(wordCounts)  # Obtiene solo las 20 palabras más frecuentes

    try:
        with open(outputPath, "w", encoding="utf-8") as f:
            f.write("WORD\tCOUNT\n")
            for word, count in top_20:
                f.write(f"{word}\t{count}\n")

        # Imprimir los resultados en pantalla
        print("\nResultados (Top 20):")
        print(f"{'WORD':<15}{'COUNT':<10}")
        print("-" * 25)
        for word, count in top_20:
            print(f"{word:<15}{count:<10}")

    except Exception as e:
        print(f"Error al escribir el archivo {outputPath} -> {e}")

def Main():
    """Función principal que ejecuta todo el proceso."""
    start_time = time.time()

    inputFiles = [
        "/content/sample_data/TC1.txt", "/content/sample_data/TC2.txt",
        "/content/sample_data/TC3.txt", "/content/sample_data/TC4.txt",
        "/content/sample_data/TC5.txt"
    ]
    outputFiles = [
        "/content/sample_data/TC1.Results.txt", "/content/sample_data/TC2.Results.txt",
        "/content/sample_data/TC3.Results.txt", "/content/sample_data/TC4.Results.txt",
        "/content/sample_data/TC5.Results.txt"
    ]

    for i in range(len(inputFiles)):  # Usa range(len()) en lugar de enumerate() (mala práctica)
        print(f"\nProcesando archivo: {inputFiles[i]}")  # Mensaje innecesario
        words = ReadFile(inputFiles[i])
        if words is None:
            print(f"Archivo {inputFiles[i]} no se procesó correctamente.")
            continue
        wordCounts = CountWords(words)
        WriteResults(outputFiles[i], wordCounts)

    exec_time = time.time() - start_time
    print(f"\nTiempo de ejecución: {exec_time:.6f} segundos")

if __name__ == "__main__":
    Main()
```

```python
import os


def read_file(file_path):
    """Lee palabras de un archivo y maneja errores en los datos."""
    words = []
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            for line in file:
                words.extend(line.strip().split())
    except FileNotFoundError as error:
        print(f"Error: No se encontró el archivo {file_path} -> {error}")
        return None
    except Exception as error:
        print(f"Error al leer el archivo {file_path} -> {error}")
        return None
    return words


def count_words(word_list):
    """Cuenta las palabras en la lista y devuelve un diccionario."""
    counts = {}
    for word in word_list:
        counts[word] = counts.get(word, 0) + 1
    return counts


def get_top_20(word_counts):
    """Obtiene las 20 palabras más frecuentes."""
    return sorted(word_counts.items(), key=lambda x: x[1], reverse=True)[:20]


def write_results(output_path, word_counts):
    """Escribe los resultados en un archivo y los imprime en pantalla."""
    top_20 = get_top_20(word_counts)
    try:
        with open(output_path, "w", encoding="utf-8") as file:
            file.write("WORD\tCOUNT\n")
            for word, count in top_20:
                file.write(f"{word}\t{count}\n")

        # Imprimir los resultados en pantalla
        print("\nResultados (Top 20):")
        print(f"{'WORD':<15}{'COUNT':<10}")
        print("-" * 25)
        for word, count in top_20:
            print(f"{word:<15}{count:<10}")

    except Exception as error:
        print(f"Error al escribir el archivo {output_path} -> {error}")


def process_files():
    """Procesa los archivos de entrada y genera los archivos de salida con el Top 20 de palabras."""
    input_files = [
        "/content/sample_data/TC1.txt", "/content/sample_data/TC2.txt",
        "/content/sample_data/TC3.txt", "/content/sample_data/TC4.txt",
        "/content/sample_data/TC5.txt"
    ]
    output_files = [
        "/content/sample_data/TC1.Results.txt", "/content/sample_data/TC2.Results.txt",
        "/content/sample_data/TC3.Results.txt", "/content/sample_data/TC4.Results.txt",
        "/content/sample_data/TC5.Results.txt"
    ]

    for input_file, output_file in zip(input_files, output_files):
        print(f"\nProcesando archivo: {input_file}")
        words = read_file(input_file)
        if words is None:
            print(f"Archivo {input_file} no se procesó correctamente.")
            continue
        word_counts = count_words(words)
        write_results(output_file, word_counts)


if __name__ == "__main__":
    process_files()
```
