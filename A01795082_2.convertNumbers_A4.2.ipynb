```python
import os, time   # Importaciones en la misma línea (mala práctica)

def ReadNumbersFromFile(filePath):   # Nombre en CamelCase (mala práctica)
    """Lee números de un archivo y maneja errores en los datos."""
    numbers = []
    try:
        file = open(filePath, "r", encoding="utf-8")  # No usa with open()
        for line in file:
            line = line.strip()
            try:
                numbers.append(int(line))
            except:
                numbers.append(None)  # No especifica la excepción
        file.close()
    except:
        return None
    return numbers

def ConvertToBinary(n):
    """Convierte un número entero a su representación binaria sin usar bin()."""
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary

def ConvertToHexadecimal(n):
    """Convierte un número entero a su representación hexadecimal sin usar hex()."""
    if n == 0:
        return "0"
    hex_digits = "0123456789ABCDEF"
    hexadecimal = ""
    while n > 0:
        hexadecimal = hex_digits[n % 16] + hexadecimal
        n //= 16
    return hexadecimal

def ProcessNumbers(inputFile, outputFile):
    """Lee un archivo, convierte los números y guarda los resultados en otro archivo."""
    start_time = time.time()
    numbers = ReadNumbersFromFile(inputFile)

    if numbers is None:
        print(f"Error: No se pudo leer {inputFile}")  # Imprime mensajes de error en la función
        return

    results = []
    for i in range(len(numbers)):  # Usa range(len()) en lugar de enumerate() (mala práctica)
        num = numbers[i]
        if num is None:
            binary, hexValue = "#VALUE!", "#VALUE!"
        else:
            binary, hexValue = ConvertToBinary(num), ConvertToHexadecimal(num)
        results.append((i + 1, num if num is not None else "ERROR", binary, hexValue))

    print("\nITEM    VALUE    BIN         HEX")  # Impresión sin formato de tabla correcto
    for row in results:
        print(row[0], row[1], row[2], row[3])

    file = open(outputFile, "w", encoding="utf-8")  # No usa with open()
    file.write("ITEM    VALUE    BIN         HEX\n")
    for row in results:
        file.write(f"{row[0]}    {row[1]}    {row[2]}    {row[3]}\n")
    file.close()

    exec_time = time.time() - start_time
    print(f"Tiempo de ejecución: {exec_time:.6f} segundos")

def ProcessAllFiles():
    """Procesa TC1.txt a TC4.txt y guarda los resultados en ConvertionResults.txt"""
    inputFolder = "/content/sample_data/"
    outputFile = os.path.join(inputFolder, "ConvertionResults.txt")
    testCases = ["TC1.txt", "TC2.txt", "TC3.txt", "TC4.txt"]

    for testCase in testCases:
        ProcessNumbers(os.path.join(inputFolder, testCase), outputFile)

# Ejecutar el procesamiento
ProcessAllFiles()
```

```python
# -*- coding: utf-8 -*-
"""convertNumbers

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/13ZbIH8H_jOqbTQcgBZe6_lxNd2AEjjJH
"""

import os

def read_numbers_from_file(file_path):
    """Lee números de un archivo y maneja errores en los datos."""
    numbers = []
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            for line in file:
                line = line.strip()
                try:
                    numbers.append(int(line))
                except ValueError:
                    numbers.append(None)  # Indica error en conversión
    except FileNotFoundError:
        return None
    return numbers

def int_to_binary(n):
    """Convierte un número entero a su representación binaria sin usar bin()."""
    if n == 0:
        return "0"
    is_negative = n < 0
    n = abs(n)
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return f"-{binary}" if is_negative else binary  # Representación simple de negativos

def int_to_hexadecimal(n):
    """Convierte un número entero a su representación hexadecimal sin usar hex()."""
    if n == 0:
        return "0"
    is_negative = n < 0
    n = abs(n)
    hex_digits = "0123456789ABCDEF"
    hexadecimal = ""
    while n > 0:
        hexadecimal = hex_digits[n % 16] + hexadecimal
        n //= 16
    return f"-{hexadecimal}" if is_negative else hexadecimal  # Representación simple de negativos

def process_numbers(input_file, output_file):
    """Lee un archivo, convierte los números y guarda los resultados en otro archivo."""
    numbers = read_numbers_from_file(input_file)

    if numbers is None:
        return

    results = []
    for idx, num in enumerate(numbers, start=1):
        if num is None:
            binary = hex_value = "#VALUE!"
        else:
            binary = int_to_binary(num)
            hex_value = int_to_hexadecimal(num)

        results.append((idx, num if num is not None else "ERROR", binary, hex_value))

    # Imprimir resultados en pantalla con formato tabla
    header = f"{'ITEM':<10} {'VALUE':<15} {'BIN':<25} {'HEX':<15}"
    print("\n" + header)
    print("-" * len(header))

    for row in results:
        print(f"{row[0]:<10} {row[1]:<15} {row[2]:<25} {row[3]:<15}")

    # Guardar resultados en archivo
    with open(output_file, "w", encoding="utf-8") as file:
        file.write(header + "\n")
        file.write("-" * len(header) + "\n")
        for row in results:
            file.write(f"{row[0]:<10} {row[1]:<15} {row[2]:<25} {row[3]:<15}\n")

    print(f"\nResultados guardados en: {output_file}")

def process_all_files():
    """Procesa TC1.txt a TC4.txt y guarda los resultados en ConvertionResults.txt"""
    input_folder = "/content/sample_data/"
    output_file = os.path.join(input_folder, "ConvertionResults.txt")
    test_cases = [f"TC{i}.txt" for i in range(1, 5)]

    for test_case in test_cases:
        input_path = os.path.join(input_folder, test_case)
        process_numbers(input_path, output_file)

# Ejecutar procesamiento
process_all_files()
```
